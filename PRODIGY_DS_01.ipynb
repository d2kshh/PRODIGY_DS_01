### Create a bar chart or histogram to visualize the distribution of a categorial or continuous variable, such as the distribution of ages or genders in a population

### Importing the necessary Libraries.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Reading the data

```python
data = pd.read_csv("World_population.csv")
```

```python
data.head()
```

```python
data.describe
```

```python
countries_to_plot = ['United States', 'Canada', 'Brazil', 'Japan', 'Germany', 'France']

# Filter data
filtered_data = data[data['Country Name'].isin(countries_to_plot)]

# Plot
plt.bar (filtered_data['Country Name'], filtered_data['2020'])
plt.xlabel ('Country Name')
plt.ylabel ('Population in 2020')
plt.title('Population of selected countries in 2020')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()
plt.show()
```
